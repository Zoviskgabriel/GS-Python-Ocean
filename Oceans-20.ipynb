```python
import pandas as pd  # Importa a biblioteca pandas para manipulação de dados
import matplotlib.pyplot as plt  # Importa a biblioteca matplotlib para visualização de dados

def coletar_dados():
  """Coleta dados do usuário e retorna um DataFrame."""
  dados = []  # Inicializa uma lista para armazenar os dados coletados
  while True:
    local = input("Digite o nome do local (ou 'sair' para finalizar): ")  # Solicita o nome do local
    if local.lower() == 'sair':  # Verifica se o usuário digitou 'sair' para finalizar a coleta
      break
    temp = input("Digite a temperatura da água (°C): ")  # Solicita a temperatura da água
    salinidade = input("Digite a salinidade (ppm): ")  # Solicita a salinidade da água
    ph = input("Digite o pH: ")  # Solicita o pH da água
    poluicao = input("Digite o nível de poluição (ppm): ")  # Solicita o nível de poluição da água
    # Adiciona os dados coletados à lista, convertendo os valores para float
    dados.append([local, float(temp), float(salinidade), float(ph), float(poluicao)])

  # Cria um DataFrame com os dados coletados e retorna
  df = pd.DataFrame(dados, columns=['Local', 'Temperatura', 'Salinidade', 'pH', 'Poluição'])
  return df

def analisar_dados(df):
  """Analisa os dados e exibe informações básicas."""
  print("Análise dos Dados:")
  # Calcula e exibe a temperatura média
  print(f"Temperatura média: {df['Temperatura'].mean():.2f} °C")
  # Calcula e exibe a salinidade média
  print(f"Salinidade média: {df['Salinidade'].mean():.2f} ppm")
  # Calcula e exibe o pH médio
  print(f"pH médio: {df['pH'].mean():.2f}")
  # Calcula e exibe o nível de poluição médio
  print(f"Nível de poluição médio: {df['Poluição'].mean():.2f} ppm")

  # Configura o tamanho do gráfico
  plt.figure(figsize=(8, 6))
  # Plota a temperatura
  plt.plot(df['Temperatura'], label='Temperatura')
  # Plota a salinidade
  plt.plot(df['Salinidade'], label='Salinidade')
  # Plota o pH
  plt.plot(df['pH'], label='pH')
  # Plota o nível de poluição
  plt.plot(df['Poluição'], label='Poluição')
  # Adiciona uma legenda ao gráfico
  plt.legend()
  # Define o título do gráfico
  plt.title('Monitoramento dos Oceanos')
  # Define o rótulo do eixo x
  plt.xlabel('Amostras')
  # Define o rótulo do eixo y
  plt.ylabel('Valores')
  # Exibe o gráfico
  plt.show()

def salvar_dados(df):
  """Salva os dados em um arquivo CSV com colunas em linhas separadas."""
  # Solicita o nome do arquivo para salvar os dados
  nome_arquivo = input("Digite o nome do arquivo para salvar (ex: dados_oceanos.csv): ")
  # Abre o arquivo para escrita
  with open(nome_arquivo, 'w') as arquivo:
    # Escreve os dados no arquivo, coluna por coluna
    for coluna in df.columns:
      arquivo.write(f"{coluna}:\n")
      for valor in df[coluna]:
        arquivo.write(f"{valor}\n")
      arquivo.write("\n")
  # Informa ao usuário que os dados foram salvos
  print(f"Dados salvos em {nome_arquivo}")

def main():
  """Função principal do sistema."""
  # Coleta os dados do usuário
  df = coletar_dados()
  # Analisa e exibe os dados coletados
  analisar_dados(df)
  # Salva os dados coletados em um arquivo CSV
  salvar_dados(df)

if __name__ == '__main__':
  main()  # Executa a função principal se o script for executado diretamente
```
